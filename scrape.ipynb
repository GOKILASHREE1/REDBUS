```python
#corrected final for APSRTC
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import csv

# Open the target website
URL = "https://www.redbus.in/online-booking/apsrtc/?utm_source=rtchometile"
driver = webdriver.Chrome()
driver.get(URL)
driver.maximize_window()

wait = WebDriverWait(driver, 20)
time.sleep(5)

route_details = []

def APSRTC_page():
    routes = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "route_link")))
    for route in routes:
        try:
            route_link = route.find_element(By.CSS_SELECTOR, ".route_details a").get_attribute("href")
            route_title = route.find_element(By.CSS_SELECTOR, ".route_details a").get_attribute("title")
            # append data to list
            route_details.append({
                "route_link": route_link,
                "route_name": route_title
            })
        except Exception as e:
            print(f"Error: {e}")
            continue

# Scrape data from each page
for page_number in range(1, 6):
    APSRTC_page()
    if page_number < 5:
        try:
            # Scroll to the pagination element to ensure it is in view
            next_page_element = wait.until(EC.element_to_be_clickable((By.XPATH, f"//div[contains(@class, 'DC_117_pageTabs') and text()='{page_number + 1}']")))
            driver.execute_script("arguments[0].scrollIntoView();", next_page_element)
            time.sleep(1)  # Give the page a moment to adjust
            next_page_element.click()
            
            # Wait for the page number to update to the next page
            wait.until(EC.text_to_be_present_in_element(
                (By.XPATH, '//div[contains(@class, "DC_117_pageTabs DC_117_pageActive")]'), str(page_number + 1)))
            time.sleep(5)  # Wait for the next page to load
        except Exception as e:
            print(f"Error navigating to page {page_number + 1}: {e}")
            break

route_links = [(entry['route_link'], entry['route_name']) for entry in route_details]

# Data storage for bus details
bus_data = []

def scrape_bus_data(route_link, route_name):
    driver.get(route_link)
    wait = WebDriverWait(driver, 20)
    # Verify the presence of the view buses button
    try:
        view_buses = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "i[class='p-left-10 icon icon-down']")))
        view_buses.click()
    except Exception as e:
        print(f"Error clicking view buses button: {e}")
    
    seen_buses = set()
    try:
        last_height = driver.execute_script("return document.body.scrollHeight") 
        
        while True:
            bus_items = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "bus-item")))
            for bus in bus_items:
                try:
                    busname = bus.find_element(By.XPATH, "//div[@class='travels lh-24 f-bold d-color']").text
                    if busname in seen_buses:
                        continue
                    seen_buses.add(busname)
                    
                    bustype = bus.find_element(By.XPATH, "//div[@class='bus-type f-12 m-top-16 l-color evBus']").text
                    departing_time = bus.find_element(By.XPATH, "//div[@class='dp-time f-19 d-color f-bold']").text
                    duration = bus.find_element(By.XPATH, "//div[@class='dur l-color lh-24']").text
                    reaching_time = bus.find_element(By.XPATH, "//div[@class='bp-time f-19 d-color disp-Inline']").text
                    star_rating = bus.find_element(By.CSS_SELECTOR, "div.rating-sec.lh-24 > div > span").get_attribute("innerHTML")
                    price = bus.find_element(By.CSS_SELECTOR, "div.column-seven.p-right-10.w-15.fl > div > div > span").get_attribute("innerHTML")
                    seats_available = bus.find_element(By.XPATH, "//div[@class='seat-left m-top-30']").text

                    bus_data.append({
                        "route_name": route_name,
                        "route_link": route_link,
                        "busname": busname,
                        "bustype": bustype,
                        "departing_time": departing_time,
                        "duration": duration,
                        "reaching_time": reaching_time,
                        "star_rating": star_rating,
                        "price": price,
                        "seats_available": seats_available
                    })
                    
                except Exception as e:
                    print(f"Error: {e}")
                    continue
            
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)

            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                break
            last_height = new_height

            time.sleep(3)
    
    except Exception as e:
        print(f"Error scraping bus data: {e}")

# Scrape data for each route link
for route_link, route_name in route_links:
    scrape_bus_data(route_link, route_name)
    time.sleep(2)

with open("newapsrtc_Route_name_page.csv", "w", encoding='utf-8-sig', newline='') as file:
    csv_writer = csv.DictWriter(file, fieldnames=["route_name", "route_link"])
    csv_writer.writeheader()
    csv_writer.writerows(route_details)

with open("bus_data.csv", "w", encoding='utf-8-sig', newline='') as file:
    csv_writer = csv.DictWriter(file, fieldnames=["route_name", "route_link", "busname", "bustype", "departing_time", "duration", "reaching_time", "star_rating", "price", "seats_available"])
    csv_writer.writeheader()
    csv_writer.writerows(bus_data)

# Close the driver
driver.quit()
```
