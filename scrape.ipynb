```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import csv
import time


# Open the target website
URL = "https://www.redbus.in/"
driver = webdriver.Chrome()
driver.get(URL)
driver.maximize_window()

# Scroll down a bit to ensure the element is in view
driver.execute_script("window.scrollTo(0, document.body.scrollHeight / 10);")

# Wait until the element is clickable and then click it
wait = WebDriverWait(driver, 10)
govt_buses = wait.until(EC.element_to_be_clickable((By.XPATH, "//div[@class='rtcName' and text()='APSRTC']")))
govt_buses.click()

# Wait for a few seconds to observe the click action and load the new page
time.sleep(5)

route_details = []

for page_number in range(1, 6):
    route_name_elements = driver.find_elements(By.CLASS_NAME, 'route_link')

    for route in route_name_elements:
        try:
            route1 = route.find_element(By.CLASS_NAME, 'route').text
            route_name_link = route.find_element(By.TAG_NAME, 'a').get_attribute('href')
            route_name_linknew = [route1, route_name_link]
            route_details.append(route_name_linknew)
        except:
            pass

    # Scroll down a bit to ensure the pagination links are in view
    driver.execute_script("window.scrollTo(0, 0.10*document.body.scrollHeight);")
    time.sleep(2)
    
    if page_number < 5:  
        try:
            next_page_element = driver.find_element(By.XPATH, f"//div[contains(@class, 'DC_117_pageTabs') and text()='{page_number + 1}']")
            driver.execute_script("arguments[0].scrollIntoView();", next_page_element)
            time.sleep(2)  
            next_page_element.click()
            time.sleep(5)  # Wait for the next page to load
        except:
            pass



with open("APSRTC_Route_name_page.csv","w", encoding = 'utf-8-sig', newline='') as file:
     csv_writer = csv.writer(file)
     header = ['ROUTE_NAME','ROUTE_LINK']
     csv_writer.writerow(header)
     csv_writer.writerows(route_details)

# Wait for a few seconds before closing
time.sleep(5)

# Close the driver
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import csv
import time

# Open the target website
URL = "https://www.redbus.in/"
driver = webdriver.Chrome()
driver.get(URL)
driver.maximize_window()

# Scroll down a bit to ensure the element is in view
driver.execute_script("window.scrollTo(0, document.body.scrollHeight / 10);")

# Wait until the element is clickable and then click it
wait = WebDriverWait(driver, 10)
govt_buses = wait.until(EC.element_to_be_clickable((By.XPATH, "//div[@class='rtcName' and text()='APSRTC']")))
govt_buses.click()

# Wait for a few seconds to observe the click action and load the new page
time.sleep(5)

route_details = []
# Wait until the route elements are visible
wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'route_link')))

# Find all route elements
route_name_elements= driver.find_elements(By.CLASS_NAME, 'route_link')


# Extract route details and navigate to each route link
for route in route_name_elements:
    try:
        # Extract route name
        route_name = route.find_element(By.CLASS_NAME, 'route').text

        # Extract route link
        route_name_link = route.find_element(By.TAG_NAME, 'a').get_attribute('href')

        # Navigate to the route link directly
        driver.get(route_name_link)

        # Wait for the route details page to load
        time.sleep(3)

        view_buses = wait.until(EC.element_to_be_clickable((By.XPATH, """/html/body/section/div[2]/div[4]/div/div[2]/div/div[2]/div[2]/div[1]/div/div[2]/div/div[4]/div[2]/i""")))
        view_buses.click()

        last_height = driver.execute_script("return document.body.scrollHeight")

        while True:
            bus_name_elements = driver.find_elements(By.XPATH, "//div[@class='travels lh-24 f-bold d-color']")
            bus_type_elements = driver.find_elements(By.XPATH, "//div[@class='bus-type f-12 m-top-16 l-color evBus']")
            dp_time_elements = driver.find_elements(By.XPATH, "//div[@class='dp-time f-19 d-color f-bold']")
            reaching_time_elements = driver.find_elements(By.XPATH, "//div[@class='bp-time f-19 d-color disp-Inline']")
            duration_elements = driver.find_elements(By.XPATH, "//div[@class='dur l-color lh-24']")
            seats_available_elements = driver.find_elements(By.XPATH, "//div[@class='seat-left m-top-16']")

            for i in range(len(bus_name_elements)):
                bus_name = bus_name_elements[i].text if i < len(bus_name_elements) else ""
                bus_type = bus_type_elements[i].text if i < len(bus_type_elements) else ""
                dp_time = dp_time_elements[i].text if i < len(dp_time_elements) else ""
                reaching_time = reaching_time_elements[i].text if i < len(reaching_time_elements) else ""
                duration = duration_elements[i].text if i < len(duration_elements) else ""
                seats_available = seats_available_elements[i].text if i < len(seats_available_elements) else ""

                route_details.append([
                    bus_name, bus_type, dp_time, reaching_time, duration, seats_available
                ])

            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)  # Wait for the page to load new content

            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                break  # Exit the loop if no new content is loaded
            last_height = new_height

        for bus_detail in buses_details:
            print(bus_detail)

        time.sleep(3)

        # Navigate back to the previous page
        driver.back()

        # Wait for the page to load and refetch the route elements
        wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'route_link')))
        route_name_elements = driver.find_elements(By.CLASS_NAME, 'route_link')
        
    except Exception as e:
        print(f"Error: {e}")
        continue


# Close the browser
driver.quit()
```
